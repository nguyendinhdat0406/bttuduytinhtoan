Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu 
vi của nó.  

```python
a,b = map(float, input().split())
print(f"Diện tích của hình chữ nhật là: {a*b}")
print(f"Chu vi của hình chữ nhật là: {(a + b)*2}")
```

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình.

```python
a, b, c = map(eval,input().split())
if a > b+c or b > a+c or c > a+b:
    print("không phải tam giác")
else:
    if  a==b==c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        if f"{a**2:.9f}" == f"{(b**2 + c**2):.9f}" or f"{b**2:.9f}" == f"{(a**2 + c**2):.9f}" or f"{c**2:.9f}" == f"{(a**2 + b**2):.9f}" :
            print("tam giác vuông cân")
        else:
            print("tam giác cân")
    elif f"{a**2:.9f}" == f"{(b**2 + c**2):.9f}" or f"{b**2:.9f}" == f"{(a**2 + c**2):.9f}" or f"{c**2:.9f}" == f"{(a**2 + b**2):.9f}" :
        print("tam giác vuông")
    else:
        print("tam giác thường ")
chuvi = a + b + b
dientich = (chuvi*(chuvi-a)*(chuvi-b)*(chuvi-c))**0.5
print(chuvi)
print(dientich)
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float, input().split())
denta = b**2 - 4*a*c
if a == 0:
    print("Nghiệm:", -c/b)
else:
    if denta > 0:
        x1 = (-b + (denta)**0.5)/(2*a)
        x2 = (-b - (denta)**0.5)/(2*a)
        print(f"{x1} \n {x2}")
    elif denta == 0:
        x = -b/(2*a)
        print(x)
    else:
        print("Vô nghiệm")
```
